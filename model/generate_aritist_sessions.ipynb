```python
from config_file import data_path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
import os
```

```python
df = pd.read_csv(os.path.join(data_path, 'timestamp_artist.csv'))
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df.head()
```

```python
# make df for each artist
artists_ids = df['id_artist'].unique()
artists_df = {}
os.makedirs(os.path.join(data_path, 'artists_sessions'), exist_ok=True)
counter = 0
for artist in artists_ids:
    artists_df[artist] = df[df['id_artist'] == artist]
    artists_df[artist].set_index('timestamp', inplace=True)
    artists_df[artist].to_csv(os.path.join(data_path, 'artists_sessions', f'{artist}.csv'))
    counter += 1
    print(f'{counter}/{len(artists_ids)}')
```
